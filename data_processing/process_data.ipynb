```python
import os

folder_names = [
    "1rover_1hiddenpoi",
    "1rover_1normalpoi",
    "1rover_2pois_1hidden_1normal",
    "1uav_1rover_1hiddenpoi",
    "1uav_1rover_1normalpoi",
    "1uav_1rover_2_pois_1hidden_1normal"
]

experiment_dirs = []
for name in folder_names:
    experiment_dirs.append(os.path.expanduser("~/influence-shaping/results/preliminary/may_16/tourn_fix/"+name))
```

```python
import csv
import pandas as pd
def load_trial(trial_dir):
    return pd.read_csv(trial_dir)
```

```python
import yaml
from pprint import pprint
import matplotlib.pyplot as plt

experiment_ind = 2
experiment_dir = experiment_dirs[experiment_ind]
print("Showing result for "+experiment_dir)
config_dir = experiment_dir + "/config.yaml"
with open(config_dir, 'r') as file:
    config = yaml.safe_load(file)

# Now load in each trial's result
trial_dirs = []
num_trials = config["experiment"]["num_trials"]
for t in range(num_trials):
    trial_dir = experiment_dir + "/trial_"+str(t)+"/fitness.csv"
    trial_dirs.append(trial_dir)

trial_df_list = []
for trial_dir in trial_dirs:
    trial_df_list.append(load_trial(trial_dir))

fig, ax = plt.subplots()

for trial_df in trial_df_list:
    hof_team_fitnesses = trial_df[" team_fitness"].to_numpy(dtype=float)
    ax.plot(hof_team_fitnesses)

plt.show()
```
