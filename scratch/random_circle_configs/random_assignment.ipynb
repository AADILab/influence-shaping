```python
from pathlib import Path
import os
base_dir = Path(os.path.expanduser('~/influence-shaping/results/01_05_2025/yabby/32_random_pairs/2x2'))
base_name = 'D-Indirect-'
def combination_to_str(combination):
    str_ = ''
    for value in combination:
        str_ += str(value)
    return str_
```
